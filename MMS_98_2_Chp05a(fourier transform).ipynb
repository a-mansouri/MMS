```python
import numpy as np
```

```python
x= np.zeros(128)
x
```

```python
x[60:68]=1
```

```python
k= np.fft.fft(x,128)
```

```python
q= np.fft.fftshift(k)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x)
plt.show()
```

```python
plt.plot(np.abs(q))
plt.show()
```

```python
x = np.zeros(128)
x[50:82]=1
k = np.fft.fft(x,128)
q=np.fft.fftshift(k)

plt.plot(x)
plt.show()

plt.plot(np.abs(q))
plt.show()
```
